```python
from statsbombpy import sb
import pandas as pd
```

```python
competitions = sb.competitions()
```

```python
matches = sb.matches(competition_id=43, season_id=106)
```

```python
wcfinal = matches[matches["competition_stage"] == "Final"]
```

```python
# .iloc[0] pega o valor da primeira linha da coluna match_id
match_id = int(wcfinal["match_id"].iloc[0])

# Agora sim, deve funcionar
wcfinal_events = sb.events(match_id=match_id)
```

```python
wcfinal_shots = wcfinal_events[wcfinal_events["type"] == "Shot"]

wcfinal_nonPenaltyShots = wcfinal_shots[wcfinal_shots["play_pattern"] != "Other"]
```

```python
import joblib

# 1. Caminho onde você salvou o arquivo
caminho_modelo = '../models/xg_pipeline_final.pkl'

# 2. Carregar o Pipeline completo
# Importante: o arquivo pre_processing.py e utils.py devem estar acessíveis
# no mesmo diretório ou no PYTHONPATH para o joblib reconstruir a classe.
model = joblib.load(caminho_modelo)

# 3. Testar o modelo com o seu DataFrame de chutes da final (df_chutes)
# O Pipeline vai fazer a geometria e o encoding automaticamente!
wcfinal_nonPenaltyShots['meu_xg'] = model.predict_proba(wcfinal_nonPenaltyShots)[:, 1]

# 4. Ver os resultados do Messi, Mbappé e cia.
cols_analise = ['player', 'minute', 'shot_statsbomb_xg', 'meu_xg', 'shot_outcome']
display(wcfinal_nonPenaltyShots[cols_analise].sort_values(by='meu_xg', ascending=False).head(15))
```
